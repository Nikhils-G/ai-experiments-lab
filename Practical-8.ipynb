# **Dimensionality Reduction using PCA**

```python

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


```

```python
digits = load_digits()
X = digits.data
y = digits.target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='rainbow', edgecolors='k', alpha=0.7)
plt.colorbar(scatter)
plt.title("PCA Visualization of Handwritten Digits Dataset", fontsize=14, fontweight='bold')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# **Conclusion**

PCA demonstrates how AI simplifies complex data without losing key patterns.
By projecting high-dimensional data into a lower-dimensional space, PCA reveals structure and relationships among samples  forming the foundation for many feature extraction and visualization techniques in ML.
