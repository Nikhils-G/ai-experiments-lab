# **Decision Tree Classification**

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt


iris = load_iris()
X, y = iris.data, iris.target


model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X, y)


plt.figure(figsize=(14, 8))
plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree Classifier on Iris Dataset", fontsize=14, fontweight='bold')
plt.show()
```

# **Conclusion**

Decision Trees represent the core AI principle of learning from rules.
Instead of memorizing data, they learn logical conditions that define classes.
This makes them intuitive, interpretable, and foundational for ensemble methods like Random Forests and Gradient Boosted Trees.
