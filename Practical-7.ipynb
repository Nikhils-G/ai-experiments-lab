# **Clustering (Unsupervised Learning)**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
X = iris.data
y_true = iris.target
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap="viridis", s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c="red", marker="X", s=200, label="Centroids")
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1],
            c=y_kmeans, cmap="viridis", s=50, label="Clusters")

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c="red", marker="X", s=200, label="Centroids")

plt.title("K-Means Clustering on Iris (PCA-reduced)")
plt.legend()
plt.show()
```

**Conclusion :**

Clustering shows how ML can find patterns without supervision. On the Iris dataset, K-Means successfully separates flowers into natural groups. This demonstrates the unsupervised learning objective: discover structure in raw data without explicit labels.
